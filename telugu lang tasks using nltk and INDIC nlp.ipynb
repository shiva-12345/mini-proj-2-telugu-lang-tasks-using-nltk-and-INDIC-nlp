# tokenizing text in telugu

```python
texts="దాని కనుగుణంగా వ్యవహరించాలి . పత్రసహిత సాక్ష్యాధారాల తో , వ్యవహారాల ను తనిఖీ చేయాలి . ఆడిట్ చేసే విధానం సంస్థ అవసరాల ను బట్టి , అంతర్గత తనిఖీన్ బట్టి , ఇంకా అనేక ఇతర విషయాల ను బట్టి మారుతూఉంటుంది . పద్దుపుస్తకాలలో ని అంకగణిత యదార్థత ను , సరిచూడటాని కి , సాక్ష్యాలు సమాచారం , వివరణలు , అవసరమౌతాయి . సాక్ష్యాలు వోచర్లరూపం లో , లేదా పత్రాల రూపం లో ఉండవచ్చు . కావలసిన సమాచారాన్ ని వివరణల ను ఆడిటర్ యాజమ"
for text in texts:
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
    print(words)
```

# pos tagging in telugu

```python
tagged = nltk.pos_tag(words)
print(tagged)
```
